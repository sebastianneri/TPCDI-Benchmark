```python
#import CSV files and create dataframes
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
#import figure from matplotlib
from matplotlib import pyplot as plt
```

```python
#Read historical data
#Read scale 3
df = pd.read_csv('C:/Users/User/Desktop/Master/ULB/INFO-H-Data_Warehouse/TPC_DI/Results/historical_load_Scale3_KNAF2D.csv')
df['scale'] = 3
df.head()
#Read scale 6
df2 = pd.read_csv('C:/Users/User/Desktop/Master/ULB/INFO-H-Data_Warehouse/TPC_DI/Results/historical_load_Scale6_KNAF2D.csv')
df2['scale'] = 6
df2.head()
#Read scale 9
df3 = pd.read_csv('C:/Users/User/Desktop/Master/ULB/INFO-H-Data_Warehouse/TPC_DI/Results/historical_load_Scale9_KNAF2D.csv')
df3['scale'] = 9
df3.head()
```

```python
#Merge dataframes for historical load
df = pd.concat([df, df2, df3], axis=0)
#remove dubplicates
df = df.drop_duplicates()
df.head()
```

```python
#Create a graph for historical load
#Y-axis: load time in seconds
#X-axis: scale
sns.set(style="whitegrid")
ax = sns.barplot(x="scale", y="create_db_time", data=df)
ax.set(xlabel='Scale', ylabel='Create DB Time (s)')
#show labels in the graph 
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.2f'), (p.get_x() + p.get_width() / 2., p.get_height()), ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')
#Insert title
plt.title('Create Database on Historical Load')
#save the graph
plt.savefig('C:/Users/User/Desktop/Master/ULB/INFO-H-Data_Warehouse/TPC_DI/Results/historical_load.png')
plt.show()
```
